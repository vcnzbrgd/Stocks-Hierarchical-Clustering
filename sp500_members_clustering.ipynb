```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(url)

# the first table contains tickers of the current S&P 500 constituents
members_table = tables[0]
list_sp500_tickers = dict(zip(list(members_table['Symbol'].replace('BRK.B', 'BRK-B').replace('BF.B', 'BF-B')),
         list(members_table['Security'])))

tickers = HierarchicalClustering(list(list_sp500_tickers.keys()))
```

```python
df = tickers.ultrametric_distance_matrix(start_date='2020-01-01')
```

```python
plot_dendrogram(df.rename(list_sp500_tickers, axis=1))
```
